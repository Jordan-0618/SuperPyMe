```python
mcdonalds_df = pd.read_csv('mcdonalds_sentiment_by_city.csv')
mcdonalds_df['chain'] = 'mcdonalds'
```

```python
starbucks_df = pd.read_csv('starbucks_sentiment_by_city.csv')
starbucks_df['chain'] = 'starbucks'
```

```python
subway_df = pd.read_csv('subway_sentiment_by_city.csv')
```

```python
subway_df['chain'] = 'subway'
```

```python
cities_array = starbucks_df['city']
```

```python
all_chains_df = pd.concat([starbucks_df, subway_df, mcdonalds_df])
```

```python
all_chains_df= all_chains_df.sort_values(by=['city'])
all_chains_df= all_chains_df.reset_index()
all_chains_df.head(25)
```

```python
i = 0
city_sent = -1
city = all_chains_df.loc[i,'city']
df_len  = all_chains_df['city'].count()
city_max = []


while i < df_len:
    if city == all_chains_df.loc[i,'city']:
        if all_chains_df.loc[i,'sentiment'] > city_sent:
            city_sent = all_chains_df.loc[i,'sentiment']
            chain = all_chains_df.loc[i,'chain']
        i= i+1
    elif city != all_chains_df.loc[i,'city']:
        city_max.append([city, city_sent,all_chains_df.loc[i,'lat'],all_chains_df.loc[i,'lng'],all_chains_df.loc[i,'population'],chain])
        city = all_chains_df.loc[i,'city']
        city_sent = -1

        
        
```

```python
df = pd.DataFrame(data=city_max, columns=["city", "sentiment", "lat", "lng","population", "chain"])
```
